## II. Set Up a Astro Project

Follow the instruction in the link: https://www.astronomer.io/docs/astro/cli/get-started-cli

#### 1. First Create a new folder

```python
mkdir local_airflow
```

#### 2. Switch to that created folder

```python
cd local_airflow
```

```python
astro dev init
```

#### 1. Check your Airflow instance

```python
astro dev ps
```

#### 2. Run your Airflow instance

```python
astro dev start
```

#### 3. Connect to the UI

user: admin

#### 4. Stop your Airflow instance 

```python
astro dev stop
```

## IV. Set up VS Code for Airflow

Until now, we have been using Airflow on the local machine.

However, as we use the Astro CLI, Airflow runs inside Docker Containers.

Thus, the purpose of this major step is to run VS Code inside Docker Container where Airflow is running (i.e., Airflow Scheduler).

#### 1. Install extension "dev containers" and enable it

#### 2. Run your Airflow Instance

```python
astro dev start
```

#### 3. Open a Remote Window in VS Code 

Go to bottom left-corner of VS code, then, Click on >< symbol

#### 4. Attach To a Running Container

Select the Airflow Scheduler Container

#### 5. Select the scheduler

This will open a new VS Code window with the Remote Window set to the scheduler you selected

#### 6. Open the folder (usr/local/airflow)

We find the same folders and files that we have in our project

#### 7. Fix Airflow Errors

If you open a DAG, you'll notice that there are some syntax mistakes.

In order to fix that, follow these steps:

* Go to the Container VS Code window
* Install Python extenion

#### 8. Disconnect VS Code from your Docker Containers

Click the Remote Window icon -> Close Remote Connection

#### 9. Stop your Airflow Instance

```python
astro dev stop
```
